```python
trends = []

for w in locations['WOEID']:
    if len(trends) > 0:
        trends = trends.append(get("https://api.twitter.com/1.1/trends/place.json", params={"id": w}))
    else:
        trends = get("https://api.twitter.com/1.1/trends/place.json", params={"id": w})
```

### google分類による判定（mecabによる名詞の抽出→タイプ分析）

```python
n_list = []
for text in trends['name']:
    node = mecabTagger.parse(text)
    df = pd.DataFrame([x.split("\t") for x in node.split("\n")])
    nauns = df[df[3].fillna('').str.contains('名詞')][0].tolist()
    for w in nauns:
        n_list.extend([[text, w]])    
        
df = pd.DataFrame(n_list,columns = ['word','wakachi'])        
```
